```python
import warnings
warnings.filterwarnings('ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
print(dt.now())
```

```python
targetSite = 'https://music.bugs.co.kr/chart'
request = requests.get(targetSite)
# print(request) # <Response [200]> => 요청 성공
html = request.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 노래 제목 크롤링
# 노래 제목은 'class' 속성이 'title'인 <p> 태그 내부에 있다.
titles = soup.findAll('p', {'class': 'title'})
# print(titles)
# print(len(titles))
for title in titles:
    # print(title.text.strip())
    # print(title.text.split('\n')[1])
    print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
# 가수 이름은 'class' 속성이 'artist'인 <p> 태그 내부에 있다.
artists = soup.findAll('p', {'class': 'artist'})
# print(artists)
# print(len(artists))
i = 0
for artist in artists:
    i += 1
    print(i, artist.text.strip().split('\n')[0])
```

```python
print(dt.now())
for i in range(100):
    title = titles[i].text.strip().split('\n')[0]
    artist = artists[i].text.strip().split('\n')[0]
    print('{0:3d}위 {1} - {2}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
file = open('bugsTOP100.txt', 'w')
# UnicodeEncodeError: 'cp949' codec can't encode character '\xf1' in position 29: illegal multibyte sequence
# 크롤링 데이터를 텍스트 파일에 저장하려는 경우 위와 같이 한글 인코딩 문제로 UnicodeEncodeError가 발생될 수 있다.
# 에러가 발생되면 open('bugsTOP100.txt', 'w', -1, 'UTF-8') 형태로 파일을 open 한다.

file.write('{} 현재 Bugs 뮤직 실시간 TOP 100\n'.format(dt.now()))
for i in range(100):
    title = titles[i].text.strip().split('\n')[0]
    artist = artists[i].text.strip().split('\n')[0]
    data = '{0:3d}위 {1} - {2}'.format(i + 1, artist, title)
    file.write(data + '\n')

file.close()
print('bugsTOP100.txt 파일에 쓰기 완료')
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 출력한다.
try:
    file = open('bugsTOP100.txt', 'r')
    # readlines() 메소드는 텍스트 파일의 내용 전체를 한꺼번에 읽어서 리스트로 리턴한다.
    lines = file.readlines()
    for line in lines:
        # print(line, end = '')
        print(line.strip())
except FileNotFoundError:
    print('bugsTOP100.txt 파일이 디스크에 없습니다.')
```
